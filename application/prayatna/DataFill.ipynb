```python
import pandas as pd
import numpy as np
import os
```

```python
wheat_market = pd.read_csv('markets/wheat_markets.csv')
rice_market = pd.read_csv('markets/rice_markets.csv')
bajra_market = pd.read_csv('markets/bajra_markets.csv')
print(wheat_market)
```

```python
wheat_states = wheat_market.State.unique()
wheat_states
```

```python
rice_states = rice_market.State.unique()
rice_states
```

```python
bajra_states = bajra_market.State.unique()
bajra_states
```

```python
wheat_market_dict = {}
for state in wheat_states:
   wheat_market_dict[state] = wheat_market[wheat_market.State == state]['Market Name'].unique()
print(wheat_market_dict)
```

```python
wheat_district_dict = {}
for state in wheat_states:
   wheat_district_dict[state] = wheat_market[wheat_market.State == state]['District Name'].unique()
print(wheat_district_dict)
```
